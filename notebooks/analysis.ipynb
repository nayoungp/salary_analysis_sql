--13. Employee Salary by Tenure Bracket

```python
df13 = pd.read_csv("result_13.csv")
df13.head()
```

Insights:
Salaries increase from New Hire to Mid-level, peaking around $98K at 5–8 years of service.

Surprisingly, the Senior bracket (8–11 years) has a slightly lower average salary than the Mid-level group, suggesting:

Potential salary stagnation beyond year 8

Influence of lower-paid long-tenured roles

The largest group is Mid-level, followed by Senior and Junior levels.

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

# 1st Y-axie: avg salary
ax1.set_xlabel('Tenure Bracket')
ax1.set_ylabel('Average Salary', color='tab:blue')
ax1.plot(df13['tenure_bracket'], df13['avg_salary'],
         color='tab:blue', marker='o', label='Avg Salary')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 2nd Y-axis: count of employees
ax2 = ax1.twinx()
ax2.set_ylabel('Employee Count', color='tab:orange')
ax2.bar(df13['tenure_bracket'], df13['employee_count'],
        alpha=0.5, color='tab:orange', label='Employee Count')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# title and display
plt.title('Average Salary and Employee Count by Tenure Bracket')
fig.tight_layout()
plt.show()
```
